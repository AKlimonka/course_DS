```python
import requests 
import pandas as pd
```

## Список объектов

```python
# Для скачивания данных по

offset_ = 0
limit_ = 1000
# Инициализируем переменные для создания списка и запуска цикла
objects_list = []
len_of_list = 1

# Запрашиваем данные до тех пор, пока не вернется пустой список
while len_of_list > 0:
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    res = requests.get(url)
    objects_data = res.json()
    objects_data.get('data').get('list')
    objects_list.extend(objects_data.get('data').get('list'))
    len_of_list = len(objects_data.get('data').get('list'))
    offset_ += limit_
```

```python
len(objects_list) # проверяем, что данные соответствуют числу объектов на сайте - 10587
```

```python
objids = [x.get('objId') for x in objects_list]
objids[:10]
```

```python
pd.set_option('display.max_columns', None)
data_df = pd.json_normalize(objects_list)
data_df.head(5)
```

```python
data_df.to_pickle('./first_task_data/objects_data.pkl')
```

```python
# Для скачивания данных по определенному объекту, нужно добавить его id в конец строки запроса
from tqdm import tqdm
detailed_objects_list = []
for id in tqdm(objids):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{id}'
    obj_data = requests.get(url).json().get('data')
    detailed_objects_list.append(obj_data)
#Скрипт выполняется около 30 минут...
```

```python
detailed_df = pd.json_normalize(detailed_objects_list)
```

# Выгружаем фото

```python
# Создаем отдельную директорию под фотки
try: 
    os.mkdir('./first_task_data/images')
except FileExistsError:
    pass
    #print('Директория уже создана')
else:
    print('Создана директория ./first_task_data/images')
```

```python
# Сначала очистим от объектов без фото

index_count = 0
for line in photos_ser:
    if len(line) == 0:
        photos_ser.drop(index_count, inplace=True)
    index_count += 1
```

```python
photos_ser[1][0].get('objRenderPhotoUrl') # вот так выглядит ссылка
```

```python
number_of_objects = 10 # Утановим ограничение, чтобы не завалить всю память фотками. 
                       # Если нужно выгрузить все объекты, то ставим убираем переменную 
                       # в т.ч. и из запроса photos_ser[:number_of_objects]/ либо можно поставить большое число, напр 1000000
                       # Фото 10 объектов занимают около 50 МБ
            
for object_ in tqdm(photos_ser[:number_of_objects]): #перебираем объекты 
    address = detailed_df[detailed_df['id'] == object_[0].get('objId')]['address'].values 
    #под каждый объект создаем свою папку вида ИД объекта_адрес
    namedir = (str(object_[0].get('objId')) +'_'+ address)[0]
    try: 
        os.mkdir(f'./first_task_data/images/{namedir}')
    except FileExistsError:
        print('Директория уже создана') # для отслеживания косяков
    #else:
        #print(f'Создана директория {namedir}') # закомментил, чтобы не перебивало прогресс-бар
        
    #по каждому объекту перебираем фото
    for pic in object_:
        url_img = pic.get("objRenderPhotoUrl")
        name_img = pic.get('objRenderPhotoNm')
        img = requests.get(url_img)
        out = open(f'./first_task_data/images/{namedir}/{name_img}', 'wb')
        out.write(img.content)
        out.close()
```
